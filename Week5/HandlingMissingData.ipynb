```python
import pandas as pd
import numpy as np

# data_preprocessing.csv

class DataPreprocessing: 
    
    def __init__(self):
        #accept file name         
        self.fname=input("enter a file name")
         # loads csv file
        self.data=pd.read_csv(self.fname)
        
    def Handling_missing_data(self):
        
        print("type of data",type(self.data))
#         print("original data is\n",self.data)

        # print sum of Na values column wise
        print("\nsum of NaN values column wise are\n",self.data.isnull().sum())
        print("\n\nsome basic statistical details\n\n",self.data.describe())
        #replacing null values by mean 
        print("\nreplacing NaN values of columns by mean values\n\n")
        self.data.replace(np.NaN,self.data.mean(),inplace=True)
        print(self.data)
    
    def Categorical_data(self):
        dummy = pd.get_dummies(self.data['Purchased'])
        dummy_country = pd.get_dummies(self.data['Country'])
        print(dummy_country)
        df=self.data.join([dummy,dummy_country])
        print(df)
    
    def Split(self):
        
        train_set=0.70*len(self.data)
        train=int(train_set)
#         print(train)
        test_set=0.30*len(self.data)
        test=int(test_set)
#         print(test)
        train_data=self.data.head(train)
        test_data=self.data.tail(test)
        print("\ntrain set\n\n",train_data)
        print("\ntest set\n\n",test_data)
      
    def Feature_Scaling(self):
        for column in self.data.columns:
            if column == 'Country'  or column == 'Purchased':    
                pass
            else:
                self.data[column] = (self.data[column] - self.data[column].mean()) / np.std(self.data[column])
        print("\n\nStandardization\n\n",self.data)
        
        for column in self.data.columns:
            if column == 'Country'  or column == 'Purchased':    
                pass
            else:
                self.data[column] =((self.data[column] - self.data[column].min()) / 
                                (self.data[column].max()-self.data[column].min()))
        print("\n\nNormalization\n\n",self.data)
        
        
    
# creates class object    
obj = DataPreprocessing()
# calling method by class object
obj.Handling_missing_data()
obj.Categorical_data()
obj.Split()
obj.Feature_Scaling()
# print(data)
# data_preprocessing.csv
```
