```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt


class Linear_Regression:
    def __init__(self):
        # loads csv file
     
        self.alpha = 0.001
        self.epoch = 1000

    def Handling_missing_data(self, dataset):

#         print("type of data",type(dataset))
#         print("original data is\n",dataset)
        # print sum of Na values column wise
#         print("\nsum of NaN values column wise are\n",dataset.isnull().sum())
#         print("\n\nsome basic statistical details\n\n",dataset.describe())
        #replacing null values by mean
#         print("\nreplacing NaN values of columns by mean values\n\n")
        dataset.replace(np.NaN, dataset.mean(), inplace=True)
        return dataset
    
    def remove_outlier(self,df):
        low = .05
        high = .95
        quant_df = df.quantile([low, high])
        for name in list(df.columns):
            if ptypes.is_numeric_dtype(df[name]):
                df = df[(df[name] > quant_df.loc[low, name]) & (df[name] < quant_df.loc[high, name])]
        return df

    
    
    def Feature_Scaling(self, train):
        for column in train.columns:
            train[column] = ((train[column] - train[column].min()) /
                             (train[column].max() - train[column].min()))
        return train
    
    def Gradient_Descent(self, train_x_data, train_y_data):
        h1 = 0
        h2 = 0
        theta0 = 0.5
        theta1 = 0
        lists=[]
        for length in range(self.epoch):
            for i in range(len(train_x_data)):
                hypothesis = theta0 + (theta1 * train_x_data[i])
                h1 += (hypothesis - train_y_data[i])
                h2 += ((hypothesis - train_y_data[i]) * train_x_data[i])
            theta0 = theta0 - ((self.alpha * h1) / len(train_x_data))
            theta1 = theta1 - ((self.alpha * h2) / len(train_x_data))
            value=theta0
            lists.append(value)
            
        return [theta0, theta1],lists

    def Test_data(self, test_x_data, list1):
#         print(list1)
        n = len(test_x_data)
        y_predict = [0] * n
        vector = np.ones(n)
        for i in range(n):
            y_predict[i] = (list1[0] + (list1[1] * test_x_data[i]))
        return y_predict
    
    def accuracy(self, y_test_data, y_predict):
        total_error = 0
        for i in range(0, len(y_test_data)):
            total_error += abs((y_predict[i] - y_test_data[i]) / y_test_data[i])
        total_error = (total_error / len(y_test_data))
        accuracy = 1 - total_error
        return accuracy * 100
    
    def plot_graph(self,list1,x_train_data,y_train_data,x_test_data,y_test_data):
        x_axes=list1[0]
        y_axes=self.epoch
#         print(x_axes,y_axes)
#         plt.plot(x_axes, y_axes,color='red')
        plt.scatter(x_train_data,y_train_data)
        
        # Set the x axis label
        plt.xlabel("x-axis")

        # Set the y axis label
        plt.ylabel("y-axis")
        plt.title("train data")

        # Show the figure
        plt.show()
    
    
    def plot_xytest_vs_xtestypredict(self,x_test_data,y_test_data,y_predict):
        plt.scatter(x_test_data,y_test_data,color="red")
        plt.plot(x_test_data,y_predict,color="green")
        plt.xlabel("x-axis")

        # Set the y axis label
        plt.ylabel("y-axis")
        plt.title("test vs ypredict")
        
        plt.show()
        
    def plot_xytrain_vs_xytrainypredict(self,x_train_data,y_train_data,y_predict):
        plt.scatter(x_train_data,y_train_data,color="red")
        plt.plot(x_train_data,y_predict,color="green")
        plt.xlabel("x-axis")

        # Set the y axis label
        plt.ylabel("y-axis")
        plt.title("train vs ypredict")
        
        plt.show()
    
#     def plot_theta0(self,lists):
#         print(len(lists))
        
#         for i in range(1,self.epoch):
            
            
            
        
        
def main():
    obj = Linear_Regression()
    # calling method by class object
    list1 = []

    train=pd.read_csv('train.csv')
    test=pd.read_csv('test.csv')   
   
    train = obj.Handling_missing_data(train)
    test = obj.Handling_missing_data(test)
    train=obj.remove_outlier(train)
    test=obj.remove_outlier(test)
    
    train  = obj.Feature_Scaling(train)
    
    x_train_data = np.array(train["x"])
    y_train_data = np.array(train["y"])

    x_test_data = np.array(test["x"])
    y_test_data = np.array(test["y"])

    
    list1,lists = obj.Gradient_Descent(x_train_data, y_train_data)
    y_predict = obj.Test_data(x_test_data, list1)
    y_predict_train=obj.Test_data(x_train_data,list1)
    acc = obj.accuracy(y_test_data, y_predict)
    print("accuracy is", acc)
    obj.plot_graph(list1,x_train_data,y_train_data,x_test_data,y_test_data)
    obj.plot_xytest_vs_xtestypredict(x_test_data,y_test_data,y_predict)
    obj.plot_xytrain_vs_xytrainypredict(x_train_data,y_train_data,y_predict_train)
#     obj.plot_theta0(lists)

if __name__ == '__main__':
    main()
```
