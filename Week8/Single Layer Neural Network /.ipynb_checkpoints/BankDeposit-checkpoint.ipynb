```python
def Feature_Scaling(df):
        for column in df.columns:
            df[column] = ((df[column] - df[column].min()) /
                             (df[column].max() - df[column].min()))
        return df
```

```python
df = Feature_Scaling(df)
df.head()
```

```python
def Split(data):
    train_set=0.70*len(data)
    train=int(train_set)
#         print(train)
    test_set=0.30*len(data)
    test=int(test_set)
        
    return train,test

train,test = Split(df)
```

```python
train_data=df.head(train)
test_data=df.tail(test)
```

```python
# Separating the output and the parameters data frame
def separate(df):
    output = df.y
    return df.drop('y', axis=1), output

x_data_train,y_data_train = separate(train_data)
```

```python
x_data_test,y_data_test=separate(test_data)
```

```python
class SingleLayerNeural:
    def __init__(self):
        # loads csv file
        self.alpha = 0.221
        self.epoch = 10000
        
    def Train(self,x_train_data, y_train_data,weights,bias):
        db=0.0
        for i in range(self.epoch):
            z = (np.dot(weights.T,x_train_data.T)+bias)
            A = (1 / (1 + np.exp(-z)))
            dz = np.subtract(A,y_train_data.T)
            
#             dw = np.dot(x_train_data.T,dz.T)
            dw = np.dot(x_train_data.T,dz.T)/len(x_train_data)
            db = np.sum(dz,axis=1,keepdims=True)
            db = db/len(x_train_data)
#             weights = (weights-(np.dot(self.alpha,dw)))
#             b = (b-(np.dot(self.alpha,db)))
            weights=np.subtract(weights,np.dot(self.alpha,dw))
            bias=np.subtract(bias,np.dot(self.alpha,db))
        return weights,bias
    
    def Test_data(self, x_test_data, weights,bias): 
        print("sgd",x_test_data.shape, weights.shape)
        z = (np.dot(x_test_data,weights)+bias)
        A = np.array(1 / (1 + np.exp(-z)))
        
        y_prediction = np.zeros((x_test_data.shape[0], 1), dtype=float)
        
        for i in range(0,len(A)):
#             print("sigmoid",A[i])
            if round(A[i][0], 2) <= 0.5:
                y_prediction[i][0] = 0
            else:
                y_prediction[i][0] = 1
        return y_prediction
    
    def Accuracy(self,y_predict,y_test_data):
        count=0
#         print(y_test_data.shape,y_predict.shape)
        for i in range(0,len(y_test_data)):
            if y_predict[i]==y_test_data[i]:
                count+=1
#         print("cnt",count)
        count=(count/len(y_test_data)*100)
        
#         print("accuracy of test data",(count/len(y_test_data))*100)
        return count

# def main():
# #     class instantiation
#     obj = SingleLayerNeural()
    
#     x_train_data = np.array(x_data_train)
#     print("x_train_data:",x_train_data.shape)
#     y_train_data = np.array(y_data_train)
#     y_train_data = y_train_data.reshape(len(y_train_data),1)
#     print("y_train_data:",y_train_data.shape)

    
#     x_test_data = np.array(x_data_test)
#     print("x_test_data:",x_test_data.shape)
#     y_test_data = np.array(y_data_test)
#     y_test_data=y_test_data.reshape(len(y_test_data),1)
#     print("y_test_data:",y_test_data.shape)
    

# #     initializing parameters
#     weights = 47
#     bias = 1
# #     bias = bias.reshape(len(bias),1)
    
#     weights = np.full((weights +  bias,1),.1)
    
# #     passing parameters to train the model
#     weights, bias = obj.Train(x_train_data, y_train_data,weights,bias)
    
# #     testing our train model to make prediction over  
#     y_predict_test = obj.Test_data(x_test_data, weights,bias)
    
# #     testing our train model to make prediction over  
#     y_predict_train = obj.Test_data(x_train_data,weights, bias)
#     for i in range(len(y_predict_train)):
#         print(y_predict_train[i])

# #     get accuracy of train by comparing predic and actual data
#     train_accuracy=obj.Accuracy(y_train_data, y_predict_train)
#     test_accuracy=obj.Accuracy(y_test_data, y_predict_test)
    
    

# #     train_accuracy,test_accuracy = obj.accuracy(y_test_data, y_predict_test,y_predict_train,y_train_data)
#     print("\nAccuracy Train:", train_accuracy)
#     print("Accuracy Test:",  test_accuracy)

# if __name__ == '__main__':
#     main()






def main():
    obj = SingleLayerNeural()
    # calling method by class object
   
    x_train_data = np.array(x_data_train)
    y_train_data = np.array(y_data_train)
    
    
    y_train_data = y_train_data.reshape(len(y_train_data),1)
        
    x_test_data = np.array(x_data_test)
    y_test_data = np.array(y_data_test)
    y_test_data = y_test_data.reshape(len(y_test_data),1)
    
    print("x_train_data",x_train_data.shape)
    print("y_train_data",y_train_data.shape) 
    print("x_test_data",x_test_data.shape)
    print("y_test_data",y_test_data.shape)  
    
#     x_train_data = np.column_stack((np.ones((x_train_data.shape[0], 1)), x_train_data))
#     print("x_train_data",x_train_data.shape)
#     x_test_data = np.column_stack((np.ones((x_test_data.shape[0], 1)), x_test_data))
#     print("x_test_data",x_test_data.shape)
    bias=1
    x_size=47
    weights = np.full((x_size+bias,1),0.1)
    print("weights",weights.shape)
    
    weights,b = obj.Train(x_train_data, y_train_data,weights,bias)
#     print("weights",weights)
    print("b",b)
    y_predict_train = obj.Test_data(x_train_data, weights,b)
    y_predict_test = obj.Test_data(x_test_data, weights,b)
    print("y_predict_train",y_predict_test)
    for i in range(len(y_predict_test)):
        print('> predicted=' + repr(y_predict_test[i]))
    
    acc_train=obj.Accuracy(y_predict_train,y_train_data)
    print("accuracy of train data=",acc_train)
    
    acc_test=obj.Accuracy(y_predict_test,y_test_data)
    print("accuracy of test data=",acc_test)
    
    
if __name__ == '__main__':
    main()
    
     
```
